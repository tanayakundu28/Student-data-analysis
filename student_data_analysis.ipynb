```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("student_data.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop(['Unnamed: 0'], axis = 1)
df.head()
```

# Gender Distribution

```python
ax = sns.countplot(x = 'Gender', data = df)
ax.bar_label(ax.containers[0])
plt.title('Gender Distribution')
plt.show()
```

<blockquote>Gender distribution is almost 50-50</blockquote>

# Impact of Parent's Education on Marks

```python
pe = df.groupby('ParentEduc').aggregate({'MathScore' : 'mean', 'ReadingScore' : 'mean', 'WritingScore' : 'mean'})
print(pe)
```

```python
heatmap = sns.heatmap(data = pe, annot = True)
plt.title("Impact of Parent's Education on Student's Marks")
plt.show()
```

<blockquote>Students whose parents have higher levels of education tend to score higher marks. </blockquote>

# Impact of Parent's Marital Status on Student's Marks

```python
pm = df.groupby('ParentMaritalStatus').aggregate({'MathScore' : 'mean', 'ReadingScore' : 'mean', 'WritingScore' : 'mean'})
print(pm)
```

```python
heatmap = sns.heatmap(data = pm, annot = True)
plt.title("Impact of Parent's Marital Status on Student's Marks")
plt.show()
```

<blockquote>Parent's Marital Status have not affected much to the Student's Scores</blockquote>

# Performance Comparison of Student Score based on Gender

```python
gen = df.groupby('Gender').aggregate({'MathScore' : 'mean', 'ReadingScore' : 'mean', 'WritingScore' : 'mean'})
print(gen)
```

```python
subject = gen.columns.tolist()
# print(subject)

female_scores = gen.iloc[0].tolist()
male_scores = gen.iloc[1].tolist()

gen_df = pd.DataFrame({'Subjects': subject, 'Male': male_scores, 'Female': female_scores})
print(gen_df)

# Reshape the data into long format
data_long = gen_df.melt('Subjects', var_name='Gender', value_name='Scores')
print(data_long)
```

```python
sns.barplot(x='Subjects', y='Scores', hue='Gender', data=data_long)
plt.title("Performance Comparison of Student Score based on Gender")
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.show()
```

<blockquote>Male students need to focus more on Reading and Writing, whereas Female students should put more effort on Maths.</blockquote>

# Impact of Weekly Study hours on Student's Scores

```python
df["WklyStudyHours"].unique()
```

```python
hours = df.groupby(["WklyStudyHours"], as_index = False).aggregate({'MathScore' : 'mean', 'ReadingScore' : 'mean', 'WritingScore' : 'mean'})
print(hours)
```

```python
new_hours = hours.melt("WklyStudyHours", var_name = 'Subject', value_name = 'Scores')
print(new_hours)
```

```python
sns.barplot(data = new_hours, x = 'Subject', y = 'Scores', hue = 'WklyStudyHours')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.title("Impact of Weekly Study hours on Student's Scores")
plt.show()
```

<blockquote>The study hours did not have huge impact on the scores of the Subjects</blockquote>
